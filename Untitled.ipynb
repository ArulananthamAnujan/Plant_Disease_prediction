# Here we are trying to implement the code model code for a one file

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input,Conv2D,Dense,GlobalMaxPooling2D
from tensorflow.keras.applications.vgg16 import VGG16
```

```python
def buildmodel():
    input_layer=Input(shape=(120,120,3))
    
    vgg=VGG16(include_top=False)(input_layer)
    
    #classification model
    f1=GlobalMaxPooling2D()(vgg)
    class1=Dense(512,activation='relu')(f1)
    class2=Dense(248,activation='relu')(class1)
    c_output=Dense(1,activation='softmax')(class2)
    
    #Bounding box for the regression model
    f2=GlobalMaxPooling2D()(vgg)
    reg1=Dense(512,activation='relu')(f2)
    reg2=Dense(248,activation='relu')(reg1)
    c_reg=Dense(4,activation='softmax')(reg2)
    
    model=Model(inputs=input_layer,outputs=[c_output,c_reg])
    return model
```

```python
plant_model=buildmodel()
```

# Define the Loss Function and Optimizer

```python
batches_per_epochs=len(train)
lr_decay=(1./0.75 -1)/batches_per_epochs
```

```python
opt=tf.keras.optimizers.Adam(learning_rate=0.0001)
```

```python
def localization_loss(y_true,y_hat):
    delta_cord=tf.reduce_sum(tf.square(y_true[:,:2]-y_hat[:,:2]))
    
    #now we have look for the actual width and height
    h_true=y_true[:,3]-y_true[:,1]
    w_true=y_true[:,2]-y_true[:,0]
    
    h_pred=y_hat[:,3]-y_hat[:,1]
    w_pred=y_hat[:,2]-y_hat[:,0]
    
    delta_size=tf.reduce_sum(tf.square(h_true-h_pred)+tf.square(w_true-w_pred))
    
    return delta_cord+delta_size
```

```python
cls_loss=tf.keras.losses.BinaryCrossentropy()
reg_loss=localization_loss
```

```python
print(cls_loss(y[0],cls))#real input followed by the predicted one
print(reg_loss(y[1],box).numpy())
```

# Now Train the neural network

```python
class Plant_Tracker(Model):
    def __init__(self,plant_tracker,**kwargs):
        super().__init__(**kwargs)
        self.model=plant_tracker
    
    def compile(self,opt,c_loss,reg_loss,**kwargs):
        super().compile(**kwargs)
        self.closs=c_loss
        self.regloss=reg_loss
        self.opt=opt
        
    def train_step(self,batch,**kwargs):
        X,y=batch
        
        with tf.GradientTape() as tape:
            cls,cords=self.model(X,training=True)
            
            #Calculate the loss
            cls_loss=self.closs(y[0],cls)/6
            reg_loss=self.regloss(tf.cast(y[1],tf.float32),cords)/6
            
            
            #Calculate the total loss
            total_loss=reg_loss+0.5*cls_loss
            
            #Now lets do the backtracking
            grad=tape.gradient(total_loss,self.model.trainable_variables)
            
        opt.apply_gradients(zip(grad,self.model.trainable_variables))
        
        return {'Total_loss':total_loss,'Class_loss':cls_loss,'Regression_loss':reg_loss}
    
    def test_step(self,batch,**kwargs):
        X,y=batch
        
        cls,cords=self.model(X,training=False)
        cls_loss=self.closs(y[0],cls)
        reg_loss=self.regloss(tf.cast(y[1],tf.float32),cords)
        total_loss=reg_loss+0.5*cls_loss
        
        return {'Total_loss':total_loss,'Class_loss':cls_loss,'Regression_loss':reg_loss}
    
    def call(self,X,**kwargs):
        return self.model(X,**kwargs)       
```

```python
model=Plant_Tracker(plant_model)
model.compile(opt,cls_loss,reg_loss)
```

```python
logs='logs'
tensorboard_callbacks=tf.keras.callbacks.TensorBoard(log_dir=logs)
```

```python
model.compile(opt,cls_loss,reg_loss)
hist=model.fit(train,epochs=5,callbacks=[tensorboard_callbacks])
```
